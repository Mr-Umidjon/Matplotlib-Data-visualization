```python
import matplotlib.pyplot as plt
import math
```

```python
def show_plot(x, y):
    """
    Given x and y values, this function will plot the graph
    Args:
        x(int): fist point
        y(int): second point
    Returns:
        None
    """
    # write your code here
    plt.scatter(x, y)
    plt.show()
```

```python
show_plot(4, 3)
```

```python
show_plot(4, 3)
```

```python
def show_plot_points_in_list(x, y):
    """
    Given x and y lists, this function will plot the graph
    Args:
        x(list): list of x points
        y(list): list of y points
    Returns:
        None
    """
    # write your code here
    plt.scatter(x, y)
    plt.show()


show_plot_points_in_list([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
```

```python
show_plot_points_in_list([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
```

```python
def plot_function(k):
    """
    plot a graph of the kx function.
    Asign the x and y values to a list
    x values are from 0 to 10, y = kx
    
    Args:
        k(int): integer value
    Returns:
        None
    """
    # write your code here
    y = []
    x = range(10)
    for i in x:
        y.append(k * i)
    plt.scatter(x, y)
    plt.show()
```

```python
plot_function(-2)
```

```python
plot_function(-2)
```

```python
def plot_function_1(k):
    """
    plot a graph of the kx^2 function.
    Assign the x and y values to a list
    x values are from -10 to 11, y = kx^2
    
    Args:
        k(int): integer value
    Returns:
        None
    """
    # write your code here
    x = range(-10, 11)
    y = []
    for i in x:
        y.append(k * i ** 2)
    plt.scatter(x, y)
    plt.show()
```

```python
plot_function_1(5)
```

```python
plot_function_1(5)
```

```python
from math import sin, pi


def plot_sin_function():
    """
    draw a graph of the sin function
    Assign the x and y values to a list
    x values are from -500 to 501, y = sin(x)
    """
    # write your code here
    x = range(-500, 501)
    y = []
    for i in x:
        y.append(sin(i * pi / 180))
    plt.plot(x, y)
    plt.show()
```

```python
plot_sin_function()
```

```python
plot_sin_function()
```

```python
from math import cos, pi
def plot_cos_function():
    """
    draw a graph of the sin function
    Assign the x and y values to a list
    x values are from -500 to 501, y = cos(x)
    """
    # write your code here
    x = range(-500, 501)
    y = []
    for i in x:
        y.append(cos(pi * i / 180))
    plt.plot(x, y)
    plt.show()
```

```python
plot_cos_function()
```

```python
plot_cos_function()
```
